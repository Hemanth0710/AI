```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

Observation:
1. rating b/w 4.5 to 4.9 -- Excellent
2. rating b/w 4.0 to 4.4 -- Very Good
3. rating b/w 3.5 to 3.9 -- Good
4. rating b/w 2.5 to 3.4 -- Average
5. rating b/w 1.8 to 2.4 -- Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
```

```python
sb.barplot(x='Aggregate rating',y='Rating count',data=rating,hue='Rating color',palette=('blue','red','orange','yellow','green','darkgreen'))
```

Observation:
1. Not rated counts are so high
2. max ratings are given from 2.5 to 3.4
