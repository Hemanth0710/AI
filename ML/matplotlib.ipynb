# matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,60,step=10)
y=np.arange(0,110,step=20)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("graph")
plt.savefig('graph.jpg')
```

```python
plt.plot(x,y,c='b')
```

```python
fig=plt.figure()
ax=fig.add_subplot(235)
```

```python
plt.plot(x,y,'o-.r')
plt.title('Graph')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'o:c')
plt.subplot(2,2,2)
plt.plot(x,y,'*-b')
```

```python
z=1+x
```

```python
plt.plot(x,z,'*')
```

```python
pi=np.pi
```

```python
a=np.arange(1,5*pi,0.1)
b=np.sin(a)
plt.plot(a,b)
```

```python
#piechart
x_pie=[200,300,500,600]
labels='A','B','C','D'
colors=['y','b','g','c']
explode=(0.1,0,0,0)
plt.pie(x_pie,labels=labels,colors=colors,explode=explode,shadow='True',autopct='%1.1f%%')
```
